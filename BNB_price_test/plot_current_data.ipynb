```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('crypto_data.csv')

df['timestamp'] = pd.to_datetime(df['timestamp'])

timestamps = []
prices = []
ma_7 = []
ma_25 = []
ma_99 = []
background_color = []

# Iterate over rows and collect points for plotting
for index, row in df.iterrows():
    timestamps.append(row['timestamp'])
    prices.append(row['price'])
    ma_7.append(row['MA_7'])
    ma_25.append(row['MA_25'])
    ma_99.append(row['MA_99'])

    # Check condition for background color change
    if row['MA_7'] > row['MA_25']:
        background_color.append('lightgreen')
    else:
        background_color.append('red')
        

plt.figure(figsize=(10, 6))

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(timestamps, prices, color='blue', label='Price')
plt.plot(timestamps, ma_7, color='orange', linestyle='-', label='MA_7')
plt.plot(timestamps, ma_25, color='green', linestyle='-', label='MA_25')
plt.plot(timestamps, ma_99, color='yellow', linestyle='-', label='MA_99')

# Add background color
for i in range(len(timestamps)):
    plt.axvspan(timestamps[i], timestamps[i] + pd.Timedelta(minutes=1), color=background_color[i], alpha=0.5)



# Customize the plot
plt.title('Price and Moving Averages')
plt.xlabel('Timestamp')
plt.ylabel('Value')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```
